US Crude Oil monthly imports quantity from 2009 - 01 to 2024 - 12

```python
# Define the API URL
url = "https://api.eia.gov/v2/crude-oil-imports/data/?frequency=monthly&data[0]=quantity&start=2009-01&end=2024-12&sort[0][column]=period&sort[0][direction]=desc&sort[1][column]=quantity&sort[1][direction]=asc&offset=0&length=5000"

# Define the API parameters
params = {
    "frequency": "monthly",
    "data": ["quantity"],
    "facets": {},  # Empty facets means all data
    "start": "2009-01",
    "end": "2024-12",
    "sort": [
        {"column": "period", "direction": "desc"},
        {"column": "quantity", "direction": "asc"}
    ],
    "offset": 0,
    "length": 5000  # Maximum number of results
}

# Replace with your EIA API key
api_key = "DyAfRezUNqRWwYfUjsX9crp1UjD9G3EYsj3CCfxJ"

# Send the GET request
headers = {"X-Params": str(params)}
response = requests.get(url, headers=headers, params={"api_key": api_key})

# Check if the request was successful
if response.status_code == 200:
    data = response.json()  # Convert response to JSON
    
    # Extract the 'data' field into a Pandas DataFrame
    if "response" in data and "data" in data["response"]:
        oil_import = pd.DataFrame(data["response"]["data"])
        print("Data successfully retrieved!")
        
        # Display first few rows
        oil_import.head()
    else:
        print("No data found in response.")
else:
    print(f"Error: {response.status_code}, {response.text}")
oil_import.head()
```
